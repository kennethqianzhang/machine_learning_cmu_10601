```python
import numpy as np
from pandas import read_excel
import pandas as pd
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
import seaborn as sns
import matplotlib.pyplot as plt
import mpl_scatter_density
from collections import Counter
```

```python
# load dataset
data = read_excel('HCOUGH.xlsx', header=None)

```

```python
data1 = data[0]
data1 = data1.dropna()

mean_data = np.mean(data1[1:])
```

```python
data2 = data[1]
data2 = data2.dropna()

mean_data2 = np.mean(data2[1:])
```

```python
for i in range(10):
    
    df = pd.DataFrame()
    x = resample(data1[1:], n_samples=int(len(data1)))

    df["Honey"] = x
    df["Index"] = 1
    
    df2 = pd.DataFrame()

    y = resample(data2[1:], n_samples=int(len(data2)))

    df2["DM"] = y
    df2["Index"] = 2
    
    plt.figure()
    plt.scatter(df["Honey"], df["Index"], s = [20*i for i in Counter(df["Honey"]).values() for j in range(i)], color = 'r')
    plt.scatter(df2["DM"], df2["Index"],s = [20*i for i in Counter(df2["DM"]).values() for j in range(i)], color = 'g')
    plt.ylim((0,3))
    plt.axvline(x = mean_data, color = "m")
    plt.axvline(x = mean_data2, color = "c")
    plt.axvline(x = np.mean(df["Honey"]), color = "r", linestyle = '-.')
    plt.axvline(x = np.mean(df2["DM"]), color = "g", linestyle = '-.')
    plt.show()
```

```python
mean_honey = []
mean_dm = []
for i in range(5000):
    
    df = pd.DataFrame()
    x = resample(data1[1:], n_samples=int(len(data1)))

    df["Honey"] = x
    df["Index"] = 1
    
    df2 = pd.DataFrame()

    y = resample(data2[1:], n_samples=int(len(data2)))

    df2["DM"] = y
    df2["Index"] = 2
    
    mean_honey.append(np.mean(df["Honey"]))
    mean_dm.append(np.mean(df2["DM"]))
plt.hist(mean_honey, bins='auto',color = 'r')
plt.hist(mean_dm, bins='auto', color = 'g')
```
