```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('2018Q4-duration.csv')
```

```python
df = df.fillna(0)
```

```python
df.head
```

```python
list1 = df.columns
```

```python
list1
```

```python
a = 0
b = 7
```

```python
for item in list1[1:]:
    x = df[['Duration', item]]
    y = x[item]/sum(x[item])
    plt.plot(x['Duration'],y,label=item.format(),scalex=True, scaley=True)
    plt.xlim(a, b)
    
plt.legend(loc='best',prop={'size':7})
my_x_ticks = np.arange(a+1, b+1, 1)
plt.xticks(my_x_ticks)
plt.show()
```

```python
for item in list1[1:]:
    x = df[['Duration', item]]
    x = x[df['Duration']>180]
    z = sum(x[item])/sum(df[item])
    plt.plot(item, sum(x[item])/sum(df[item]),'ro')
    plt.text(item, sum(x[item])/sum(df[item]),'%f' % z)
    print(item, ':', sum(x[item])/sum(df[item]))

```

```python

for item in list1[1:]:
    t1 = []
    t2 = []
    for i in range(int(180/30+1)):
        x = df[(df['Duration']>(i-1)*30) & (df['Duration']<=i*30)]
        z = sum(x[item])/sum(df[item])
        t1.append(i*30)
        t2.append(z)
    plt.plot(t1,t2,label=item.format())
    
plt.legend(loc='best',prop={'size':7})
my_x_ticks = np.arange(0, 180+1, 30)
plt.xticks(my_x_ticks)
    
```

```python

for item in list1[1:]:
    t1 = []
    t2 = []
    for i in range(int(28/7+1)):
        x = df[(df['Duration']>(i-1)*7) & (df['Duration']<=i*7)]
        z = sum(x[item])/sum(df[item])
        t1.append(i*7)
        t2.append(z)
    plt.plot(t1,t2,label=item.format())
    
plt.legend(loc='best',prop={'size':7})
my_x_ticks = np.arange(0, 28+1, 7)
plt.xticks(my_x_ticks)
```

```python
for item in list1[1:]:
    x = df['Duration']
    y = df[item]/sum(df[item])
    plt.plot(x,y,label=item.format(),scalex=True, scaley=True)
    plt.xlim(180, x.max())
    plt.ylim(0,0.02)
    
plt.legend(loc='best',prop={'size':7})
my_x_ticks = np.arange(180, x.max(), 30)
plt.xticks(my_x_ticks)
plt.show()
```
